# Explainable AI — SHAP, LIME, Feature Importance (Phase 6)

For every dataset, every shared output target, and every Top-3 model, this notebook generates the full explainability suite:

* **SHAP** — summary plot, beeswarm, bar feature importance, per-feature dependence plots, local waterfall plots (2 samples), and the feature-interaction matrix (tree models only).
* **LIME** — local explanations for 2 selected samples.
* **Permutation importance** — model-agnostic, on the full fitted pipeline.
* **Tree-based feature importance** — impurity-based, for tree models only.

Every individual figure and table is generated and saved under `results/xai/<dataset>/<model>/<target>/`. To keep this notebook a manageable size (hundreds of PNGs would otherwise be embedded), the SHAP/LIME figures are displayed inline only for one representative target per model per dataset — the full importance tables (permutation and tree-based) are always displayed and saved for **every** dataset/target/model combination, and the last cell shows the complete combined results.

Setup: repo imports, pandas display options, load the Top-3 models and all datasets.

```python
import sys
from pathlib import Path

REPO_ROOT = Path.cwd().parent if (Path.cwd() / '..' / 'src').resolve().exists() else Path.cwd()
SCRIPTS_DIR = REPO_ROOT / 'scripts'
for p in (REPO_ROOT, SCRIPTS_DIR):
    if str(p) not in sys.path:
        sys.path.insert(0, str(p))

import pandas as pd
from IPython.display import Image, display

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 250)

from _pipeline_common import read_top3_models
from src.config import get_base_seed, get_path, load_config
from src.data.loader import load_all_datasets
from src.models.registry import build_model
from src.utils.io import save_table
from src.xai.importance import compute_permutation_importance, tree_feature_importance
from src.xai.lime_analysis import lime_explain_samples
from src.xai.shap_analysis import generate_all_shap_plots

cfg = load_config()
seed = get_base_seed()
top3 = read_top3_models()
bundles = load_all_datasets(discrete=False)
print('Top-3 models explained:', top3)
```

## Explainability on Dataset_0136

SHAP + LIME + permutation + tree importance for every shared target and every Top-3 model. Figures for the first target of each model are shown inline as a worked example; every combination's figures are saved to disk and its importance tables are collected below.

```python
perm_rows_0136 = []
tree_rows_0136 = []
bundle = bundles['Dataset_0136']
shared_targets = [c for c in cfg['shared_outputs'] if c in bundle.Y.columns]
shown_models_0136 = set()

for target in shared_targets:
    y = bundle.Y[target]
    for model_name in top3:
        pipe = build_model(model_name, seed=seed)
        pipe.fit(bundle.X.to_numpy(), y.to_numpy())

        safe_target = target.replace('/', '_').replace(' ', '_')
        out_dir = get_path('results_dir') / 'xai' / 'Dataset_0136' / model_name / safe_target

        generate_all_shap_plots(pipe, bundle.X, out_dir)
        lime_explain_samples(pipe, bundle.X, out_dir, sample_indices=[0, 1])

        perm = compute_permutation_importance(pipe, bundle.X, y, seed=seed)
        save_table(perm, out_dir / 'permutation_importance.csv')
        perm_tagged = perm.copy()
        perm_tagged.insert(0, 'dataset', 'Dataset_0136')
        perm_tagged.insert(1, 'target', target)
        perm_tagged.insert(2, 'model', model_name)
        perm_rows_0136.append(perm_tagged)

        try:
            tree_imp = tree_feature_importance(pipe, bundle.feature_names)
            save_table(tree_imp, out_dir / 'tree_feature_importance.csv')
            tree_tagged = tree_imp.copy()
            tree_tagged.insert(0, 'dataset', 'Dataset_0136')
            tree_tagged.insert(1, 'target', target)
            tree_tagged.insert(2, 'model', model_name)
            tree_rows_0136.append(tree_tagged)
        except TypeError:
            pass

        if model_name not in shown_models_0136:
            shown_models_0136.add(model_name)
            print(f'--- Worked example: {model_name} / {target} ---')
            for fig_name in ('shap_summary', 'shap_beeswarm', 'shap_feature_importance'):
                fig_path = out_dir / f'{fig_name}.png'
                if fig_path.exists():
                    display(Image(filename=str(fig_path)))
            lime_path = out_dir / 'lime_sample_0.png'
            if lime_path.exists():
                display(Image(filename=str(lime_path)))

        print(f'[xai] Dataset_0136 / {target} / {model_name} done')

perm_0136 = pd.concat(perm_rows_0136, ignore_index=True)
tree_0136 = pd.concat(tree_rows_0136, ignore_index=True) if tree_rows_0136 else pd.DataFrame()
print()
print('Permutation importance — Dataset_0136 (all targets/models):')
display(perm_0136)
if not tree_0136.empty:
    print('Tree-based importance — Dataset_0136 (all targets/models):')
    display(tree_0136)
```

## Final Summary — Complete XAI Importance Results (All Datasets)

Permutation and tree-based feature importance combined across every dataset, target and Top-3 model, saved to `results/xai/permutation_importance_summary.csv` and `results/xai/tree_importance_summary.csv`. Every SHAP/LIME figure generated in this run is saved under `results/xai/<dataset>/<model>/<target>/`.

```python
perm_all = pd.concat([perm_0136, perm_0172, perm_3772], ignore_index=True)
tree_frames = [t for t in (tree_0136, tree_0172, tree_3772) if not t.empty]
tree_all = pd.concat(tree_frames, ignore_index=True) if tree_frames else pd.DataFrame()

xai_out_dir = get_path('results_dir') / 'xai'
save_table(perm_all, xai_out_dir / 'permutation_importance_summary.csv')
if not tree_all.empty:
    save_table(tree_all, xai_out_dir / 'tree_importance_summary.csv')

print('=' * 90)
print('1) PERMUTATION IMPORTANCE — all datasets / targets / Top-3 models')
print('=' * 90)
display(perm_all)

if not tree_all.empty:
    print()
    print('=' * 90)
    print('2) TREE-BASED FEATURE IMPORTANCE — all datasets / targets / tree models')
    print('=' * 90)
    display(tree_all)

print()
print('Mean permutation importance by feature (averaged over datasets/targets/models):')
display(perm_all.groupby('feature')['importance_mean'].mean().sort_values(ascending=False))

print()
print('All XAI figures and tables saved to:', xai_out_dir.resolve())
```
